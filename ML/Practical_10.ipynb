Write a program to implement (i) Pearson correlation and display heatmap (ii) chi-square test

```
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import math
from scipy.stats import chisquare, chi2
```

```
# Pearson's Correlation
data = np.random.rand(10,12)
```

```
PCC=np.corrcoef(data.T)
print(PCC)
```

```
heatmap = sns.heatmap(PCC, annot=True, fmt='.1f')
plt.show()
```

```
# Chi Square Test

# taking random observed and expected values
observed = [20, 50, 30, 25, 60]
expected = [10, 15, 35, 55, 40]

alpha = 0.05 # level of significance
dof = 4 # degrees of freedom

chi_square = chisquare(observed, expected)
z = chi_square.statistic
print('Chi Square Statistic value : ',z)
```

```
# finding the confidence interval
critical_point = chi2.ppf(1-alpha, dof)
if z>=critical_point:
  print('Null Hypothesis is rejected!')
else:
  print('Null Hypothesis is accepted!')
```
