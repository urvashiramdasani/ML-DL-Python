```
# Load the libraries

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```
# Load Data

data = pd.read_csv('Training_set.csv')
data.head()
```

```
data = data.drop('Unnamed: 0', axis = 1)
data.head()
```

```
# Data Preprocessing

X = data.drop(['respondent_id', 'h1n1_vaccine', 'seasonal_vaccine'], axis = 1)
X.head()
```

```
# Load Test Data

X_test = pd.read_csv('Test_set.csv')
```

```
X_test.head()
```

```
# Test Data Preprocessing

X_test['age_group'] = X_test['age_group'].map(ages)
X_test['education'] = X_test['education'].map(edu)
X_test['race'] = X_test['race'].map(race)
X_test['sex'] = X_test['sex'].map(sex)
X_test['income_poverty'] = X_test['income_poverty'].map(income)
X_test.marital_status = X_test.marital_status.map(mar)
X_test.rent_or_own = X_test.rent_or_own.map(roo)
X_test.employment_status = X_test.employment_status.map(employ)
X_test['hhs_geo_region'] = X_test['hhs_geo_region'].map(hhs)
X_test.census_msa = X_test.census_msa.map(msa)
X_test.employment_industry = X_test.employment_industry.map(indus)
X_test.employment_occupation = X_test.employment_occupation.map(occ)
```

```
for x in X_test.columns:
  X_test[x] = X_test[x].fillna(X_test[x].mode()[0])
```

```
X_test = X_test.drop(['Unnamed: 0', 'respondent_id'], axis = 1)
```

```
X_test.head()
```

```
from sklearn.linear_model import LogisticRegression
```

```
# Train the model and predict

clf = LogisticRegression(max_iter = 200)
clf.fit(X, y['h1n1_vaccine'])
```

```
y_predicted_1 = clf.predict(X_test)
```

```
clf.fit(X, y['seasonal_vaccine'])
```

```
y_predicted_2 = clf.predict(X_test)
```

```
t = pd.read_csv('Test_set.csv')
```

```
csv = np.empty((len(X_test), 3))
for i in range(len(X_test)):
  csv[i][0] = t['respondent_id'][i]
  csv[i][1] = y_predicted_1[i]
  csv[i][2] = y_predicted_2[i]
```

```
csv_file = pd.DataFrame(csv, columns = ['respondent_id', 'h1n1_vaccine', 'seasonal_vaccine'])
```

```
# Saving Predictions

csv_file.to_csv('Predictions.csv')
```
