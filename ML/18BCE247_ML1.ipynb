```
import pytesseract as tess

try:

    import Image

except ImportError:

    from PIL import Image

 #Changing the directory as the pytesseract library is only wrapper and the real engine has to be referenced seperately

tess.pytesseract.tesseract_cmd = r'/usr/bin/tesseract'

img = Image.open(r'test.png')

print(img)

text = tess.image_to_string(img)

print(text)
```

```
# PDF to image to Character
from pdf2image import convert_from_path

images = convert_from_path('sample.pdf', 1000)

for i in images:
    i.save('output.jpg', 'JPEG')
```
