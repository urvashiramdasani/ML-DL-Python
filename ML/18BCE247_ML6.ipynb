```
# Load the libraries

import numpy
import matplotlib.pyplot as plt
from sklearn import datasets, metrics
import seaborn as sns
from sklearn.tree import DecisionTreeClassifier
import pandas as pd
from sklearn import tree
```

```
# Load dataset

X, y = datasets.load_iris(return_X_y = True)
print(X.shape)
print(y.shape)
```

```
# Train test split without sklearn

X_train = X[range(0, 150, 2), :]
y_train = y[range(0, 150, 2)]
print(X_train.shape)
print(y_train.shape)
```

```
X_test = X[range(1, 150, 2), :]
y_test = y[range(1, 150, 2)]
print(X_test.shape)

print(y_test.shape)
```

```
# Fit the model - Using criterion='entropy'

classifier = DecisionTreeClassifier(criterion = 'entropy')
classifier.fit(X_train, y_train)
```

```
# Predict the values

y_predicted = classifier.predict(X_test)
print(y_predicted)
```

```
# Accuracy and other measures

print('Accuracy Score : ', metrics.accuracy_score(y_true = y_test, y_pred = y_predicted))
```

```
print('Classification Report : ')
print(metrics.classification_report(y_true = y_test, y_pred = y_predicted))
```

```
print('Confusion Matrix : ')
cm = metrics.confusion_matrix(y_true = y_test, y_pred = y_predicted)
print(cm)
sns.heatmap(cm, annot = True, cmap = 'plasma')
```

```
# Fit the model - Using criterion='gini'

classifier = DecisionTreeClassifier()
classifier.fit(X_train, y_train)
```

```
# Predict the values

y_predicted = classifier.predict(X_test)
print(y_predicted)
```

```
data = datasets.load_iris()
print(data)
```

```
df = pd.DataFrame(data = data.data, columns = data.feature_names)
df['target'] = data.target
```

```
df.head()
```

```
clf = DecisionTreeClassifier(max_depth = 4, criterion = 'entropy')
clf.fit(df[data.feature_names], df['target'])
```

```
fig, axes = plt.subplots(nrows = 1, ncols = 1, figsize = (4,4), dpi=300)
tree.plot_tree(clf, filled = True, feature_names = data.feature_names, class_names = data.target_names)
```

```
clf = DecisionTreeClassifier(max_depth = 2)
clf.fit(df[data.feature_names], df['target'])
```
