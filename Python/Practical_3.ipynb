```
import random
import numpy as np
import math
import matplotlib.pyplot as plt
```

```
values = np.random.uniform(low=0,high=100,size=500)
#print(type(values))
mean = np.mean(values)
print(mean)
std = np.std(values)
print(std)
var = math.sqrt(std)
print(var)
```

```
num_bins = int(input('Enter the bins : '))
values.sort()
y = ((1 / (np.sqrt(2 * np.pi) * std)) *
     np.exp(-0.5 * (1 / std * (values - mean))**2))
plt.plot(values,y,'-')
plt.axvline(mean,color="red",linewidth=2,label="mean")
plt.axvline(std,color="black",linewidth=2,label="std")
plt.legend() #to display label
#plt.plot(mean,y,'-')
```

```
z_score = []
for value in values:
  z_score.append((value-mean)/std)

#y_pos = np.arange(bins)
#fig, ax = plt.subplots()
plt.hist(values, num_bins, edgecolor='black')
plt.title('Histogram of values')
plt.xlabel('Values')
plt.ylabel('Count')
```
