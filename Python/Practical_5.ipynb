Write a program which scans marks of 200 students in P & S. Assuming that the students are to be graded on 5 grade-scale (A, B, C, D, and F), display marks and grade of each student. Fit Gaussian distribution to the data. Repeat the exercise if students are to be graded on 7 grade-scale (A, B, C+, C, C-, D, F).

```
import random
import numpy as np
import matplotlib.pyplot as plt
```

```
marks = []
for i in range(0,200):
  marks.append(random.randint(0,100))
mean = np.mean(marks)
std = np.std(marks)
z_scores = []

for x in marks:
  z_scores.append((x-mean)/std) # cannot use directly marks - mean....

print(z_scores)
#grade = np.array([0,0,0,0,0],int)
grade = []
count_a = 0
count_b = 0
count_c = 0
count_d = 0
count_f = 0

for z in z_scores:
  if z<=-2.0:
    grade.append('F')
    count_a +=1
  elif z<=-0.5 and z>-2:
    grade.append('D')
    count_b +=1
  elif z<=0.5 and z>-0.5:
    grade.append('C')
    count_c +=1
  elif z<=2 and z>0.5:
    grade.append('B')
    count_d +=1
  elif z>2:
    grade.append('A')
    count_f +=1
#grade.sort()
for i in range(0,200):
  print(str(i)+'\t'+str(marks[i])+'\t'+str(grade[i]))

num_students = [count_a,count_b,count_c,count_d,count_f]
print(num_students)

plt.hist(grade,bins=5,edgecolor="black")
plt.title('Grade vs Number of Students')
plt.xlabel('Grades')
plt.ylabel('Number of Students')
```

```
marks.sort()
#print(marks)
sorted_z = []

for x in marks:
  sorted_z.append((x-mean)/std)
#print(sorted_z)
for i in range(0,200):
  print(str(marks[i])+'\t'+str(sorted_z[i]))
```

```
grades = []

for z in z_scores:
  if z<=-1.2:
    grades.append('F')
  elif z<=-0.6 and z>-1.2:
    grades.append('D')
  elif z<=0.09 and z>-0.6:
    grades.append('C')
  elif z<=.39 and z>0.09:
    grades.append('B')
  elif z>.39 and z<=0.88:
    grades.append('B+')
  elif z<=1.47 and z>.88:
    grades.append('A')
  elif z<=1.68 and z>1.47:
    grades.append('A+')
print(grades)
plt.hist(grades,bins=7,edgecolor="black")
plt.title('Using 7 Grades')
plt.xlabel('Grades')
plt.ylabel('Number of Students')
```

```
const = 1/(std * np.sqrt(2*np.pi))
f = []
for i in sorted_z:
  f.append(const * np.exp(-0.5 * i**2))

plt.plot(marks, f)
plt.title('Gaussian Distribution of Data')
plt.xlabel('Marks')
plt.ylabel('Variation of number of students')
plt.show()
```
