Q 9 A Develop a python program that reads the image, display matrix representation of an image creates a histogram of the image and apply the smoothing effect on an image.

```
import cv2
from google.colab.patches import cv2_imshow # Google colab doesn't support cv2.imshow()
import matplotlib.pyplot as plt
```

```
img = cv2.imread('image.jpg',0)
cv2_imshow(img)
print(img)
print(type(img))
print(img.shape)
```

```
# number of bins = 256, range = (0,256)
plt.hist(img.ravel(), 256, (0,256)) # img.ravel() returns 1d array of image
plt.show()
```

```
# different image filters
# averaging - normal blur

blur_img = cv2.blur(img, (5,5))
cv2_imshow(blur_img)
```

```
# Gaussian Filtering - most effective in removing noise

gaussian_img = cv2.GaussianBlur(img, (5,5), 0)
cv2_imshow(gaussian_img)
```

```
# Median Filtering - used to remove salt pepper noise

median_image = cv2.medianBlur(img, 5)
cv2_imshow(median_image)
```

```
# Bilateral Filtering - retains the edges

bilateral_image = cv2.bilateralFilter(img,9,75,75)
cv2_imshow(bilateral_image)
```
